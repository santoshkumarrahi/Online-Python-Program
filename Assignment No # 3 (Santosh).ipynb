```python
#4. Track passengers on a bus:
passengers = ["Ali", "Zara", "Ahmed"]
passengers.insert(2,"Sara")
print("Insert a new passenger Sara at position 2 : ",passengers)
passengers.remove("Zara")
print("Remove passesnger who got off Zara : ",passengers)
passengers.reverse()
print("Reverse the order to show who entered first : ",passengers)
```
